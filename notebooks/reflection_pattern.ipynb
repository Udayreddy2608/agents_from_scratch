```python
import os
from pprint import pprint
from groq import Groq
from dotenv import load_dotenv
from IPython.display import display_markdown

load_dotenv()

client = Groq()
```

```python
generation_chat_history = [
    {
        "role":"system",
        "content": "You are a python programmer tasked with generating high quality python code."
        "Your task is to generate the best content possible for the user's request. If the user provides critique,"
        "respond with a revised version of your previous attempt."
    }
]
```

```python
generation_chat_history.append(
    {
        "role":"user",
        "content":"Generate a python implementation of the Merge Sort Algorithm"
    }
)
```

```python
merge_sort_code = client.chat.completions.create(
    messages= generation_chat_history,
    model= "llama-3.1-8b-instant"
).choices[0].message.content

generation_chat_history.append(
    {
        "role": "assistant",
        "content":merge_sort_code
    }
)
```

```python
display_markdown(merge_sort_code, raw = True)
```

```python
reflection_chat_history = [
    {
        "role":"system",
        "content": "you are an experienced computer scientist. You are tasked with generating critique and recommendations for the user's goal"
    }
]
```

```python
reflection_chat_history.append(
    {
        "role":"user",
        "content": merge_sort_code
    }
)
```

```python
critique = client.chat.completions.create(
    messages= reflection_chat_history,
    model= "llama-3.1-8b-instant"
).choices[0].message.content
```

```python
display_markdown(critique, raw = True)
```

```python
generation_chat_history.append(
    {
        "role":"user",
        "content": critique
    }
)
```

```python
essay = client.chat.completions.create(
    messages= generation_chat_history,
    model= "llama-3.1-8b-instant"
).choices[0].message.content
```

```python
display_markdown(essay, raw = True)
```

```python
from reflection_agent import ReflectionAgent
```

```python
agent = ReflectionAgent()
```

```python
generation_prompt = """
You are a senior Python engineer.
Write clean, correct, production-ready Python code.

Rules:
- Follow PEP8
- Use type hints and docstrings
- Handle edge cases
- Keep it modular and readable
- Output only Python code
"""

reflection_prompt = """
You are a Python code reviewer.
Review the generated code for correctness, edge cases, readability,
performance, and best practices.

If issues exist:
- Fix them
- Return the improved full code

If correct:
- Return the original code unchanged

Output only Python code.
"""

user_prompt = """
Generate a python implementation of merge sort algorithm in python
"""
```

```python
final_response = agent.run(
    generation_system_prompt= generation_prompt,
    reflection_system_prompt= reflection_prompt,
    user_prompt= user_prompt,
    n_steps= 3,
    verbose= 1
)
```

```python
display_markdown(final_response, raw = True)
```
